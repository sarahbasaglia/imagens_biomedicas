```python
import numpy as np
import matplotlib.pyplot as plot
import cv2
import skimage
import skimage.exposure
```

##### Ler a imagem, normalizar e plotar a imagem 


```python
image = cv2.imread(r"C:\Unifesp\4_termo\imagens_biomedicas\aula1_introducao\raioXTorax.pgm",0)

norm = skimage.img_as_float(image)

plot.figure(0)
plot.imshow(norm,cmap = 'gray')
plot.title('Raio X')
plot.colorbar()
plot.show()
```

#### Obter dimenso da imagem 

```python
(M,N) = np.shape(norm)
print(f'Valor: ',(M,N))
```

#### Obter valores especficos 

```python
maximo = np.max(norm)
minimo = np.min(norm)
media = np.mean(norm)
dP = np.std(norm)

print(f'Mximo: ', maximo)
print(f'Mnimo: ', minimo)
print(f'Mdia: ', media)
print(f'DP: ', dP)
```
